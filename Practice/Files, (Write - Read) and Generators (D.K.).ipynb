1) Прога, которая скопирует исходный код с одного файла и вставит его в другой. (В нашем случае файлы first.py и second.py в нужной директории).    
ENG: Write a Python program to create a copy of its own source code.

```python
# Такой вариант с проверкой
import os

def file_copy(first, second):
    if not os.path.exists(first):
        print(f"Файл {first} не существует.")
        return

    try:
        with open(first, 'rb') as f1, open(second, 'wb') as f2:
            f2.write(f1.read())
        print(f"Файл успешно скопирован в {second}")
    except OSError as e:
        print(f"Ошибка при копировании: {e}")

first = "C:\\Users\\david\\Desktop\\first.py"
second = "C:\\Users\\david\\Desktop\\second.py"
file_copy(first, second)
```

4) Прога, которая считает количество строк в файле.   

```python
def count_lines(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        return sum(1 for _ in f)
```

5) Прога, которая считает количество слов в файле.

```python
def count_words(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        return sum(len(line.split()) for line in f)
```

6) Прога, которая считает количество символов в файле.

```python
def count_chars(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        return sum(len(line) for line in f)
```

7) Прога, которая выведет нам строки, начинающиеся на "А".

```python
def lines_starting_with_A(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        return [line.strip() for line in f if line.startswith("A")]
```

8) Прога, которая выведет нам последние 10 строк в файле.

```python
from collections import deque

def last_n_lines(filename, n=10):
    with open(filename, 'r', encoding='utf-8') as f:
        return list(deque(f, maxlen=n))
```

9) Генератор, который выведет по N строк из файла.

```python
def read_in_blocks(filename, n):
    block = []
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            block.append(line.rstrip('\n'))
            if len(block) == n:
                yield block
                block = []
        if block:
            yield block
```

10) Генератор по символам

```python
def chars_generator(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        while True:
            c = f.read(1)
            if not c:
                break
            yield c
```

11) Генератор по словам, без стоп слов.   

```python
STOPWORDS = {"и", "в", "на", "по", "с"}

def words_no_stop(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            for word in line.split():
                if word.lower() not in STOPWORDS:
                    yield word
```

12) Генератор уникальных слов

```python
def unique_words(filename):
    seen = set()
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            for word in line.split():
                if word not in seen:
                    seen.add(word)
                    yield word
```

13) Генератор с паузой каждые N строк. 

```python
def pause_every_n(filename, n=3):
    count = 0
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            yield line.rstrip('\n')
            count += 1
            if count % n == 0:
                yield
```

14) Генератор строк в обратном порядке без загрузки всего файла.

```python
def reverse_lines(filename):
    with open(filename, 'rb') as f:
        f.seek(0, 2)  # конец файла
        pos = f.tell()
        line = b''
        while pos > 0:
            pos -= 1
            f.seek(pos)
            char = f.read(1)
            if char == b'\n':
                yield line[::-1].decode()
                line = b''
            else:
                line += char
        if line:
            yield line[::-1].decode()
```

15) Надо написать генератор, который читает переданный по имени файл по словам. `rfbw = read_file_by_word('file.txt'). print(next(rfbw)) - получаем одно слово. print(rfbw.send(5)) - получаем 5 слов. for word in rfbw: print(word) - получаем по одному слову`.   

```python
def read_file_by_words(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        words = f.read().split()  # разбиваем весь текст на слова
        i = 0  # текущий индекс

        while i < len(words):
            n = (yield words[i])  # отдаём слово, ждём send()
            if n is None:
                # обычный режим — одно слово
                i += 1
            else:
                # режим send(N) — отдаём N слов списком
                n = int(n)
                out = words[i:i+n]
                i += n
                # yield сразу отдаёт список
                n = (yield out)


rfbw = read_file_by_words('file.txt')

print(next(rfbw))   # одно слово
print(rfbw.send(5)) # пять слов списком
print(next(rfbw))   # снова одно слово после send

for w in rfbw:
    print(w)        # слова по одному
```
