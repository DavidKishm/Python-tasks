4) В работе с `enumerate`, когда у нас есть список например и мы должны проиндексировать слова из списка, `{}` - будут для словаря где ключ это слово, а value это индекс.   

```python
def list_to_dict(lst):
    return {key:value for value,key in enumerate(lst, start=1)}

lst = ['Barca','Arsenal','Inter','PSG']

list_to_dict(lst)
```

7) Когда у нас есть например список из цифр и нам нужно часто удалять или добавлять элементы с обеих сторон или с шагом, лучше всегда импортировать для этого модуль `deque` из `collections`.   
(FIFO, LIFO, обходы и сдвиги, повороты rotate) Для всего этого deque идеально. Так код работает быстрее.      

```python
from collections import deque

d = deque([1, 2, 3])
d.append(4)       # [1, 2, 3, 4]
d.popleft()       # 0 → [1, 2, 3]
d.rotate(1)       # [3, 1, 2]
```

13) Что касается словарей, хотелось бы отметить ключевые и основные моменты работы с ними.   

```python
# Так как dict не имеет append как list или add как у set, со словарями работать чуть запутаннее.
from collections import defaultdict   # defaultdict мы импортируем во избежение ошибки в связи с тем, что у нас нет значения. То есть Python не знает к чему добавить ключ

def povtor(nums):
    d = defaultdict(int)               # int по умолчанию даёт 0
    for num in nums:
        d[num] += 1
    return dict(d)

povtor([1,2,3,1])                      # Вывод будет {1: 2, 2: 1, 3: 1} и нам не нужно ставить условие если d[num] есть в словаре, так как модуль создает ключ если его нет и не возвращает ошибку.


def povtor(nums):
    d = {}
    for num in nums:
        if num in d:
            d[num] += 1                # В данном случае ключ числа num увеличится на 1. Надо помнить что в словаре все ключи должны быть уникальны. 
        else:
            d[num] = 1
    return d

povtor([1,2,3,4,1,1])
```
